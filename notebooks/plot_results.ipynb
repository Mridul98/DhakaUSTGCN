```python
def rmse(gt,predicted):

    gt_vector = gt.values
    predicted_vector = predicted.values
    rmse = np.sqrt(np.sum((gt_vector-predicted_vector)**2)/gt_vector.shape[0])
    return rmse

def mae(gt,predicted):

    gt_vector = gt.values
    predicted_vector = predicted.values
    mae = np.sum(np.abs(predicted_vector-gt_vector))/gt_vector.shape[0]
    return mae
```

```python
def plot_graph(dataframe,day,road_number,road_tier_dict):
    title_name = "prediction_after_15_minutes_Road_{}_tier_{}_day_{}".format(road_number,road_tier_dict[int(road_number)],day)
    filters = (dataframe['day'] == day)
    road_predicted = str(road_number)
    road_gt = str(road_number)+'l'
    filtered_df = dataframe[filters]
    RMSE = rmse(filtered_df[road_gt],filtered_df[road_predicted])
    MAE = mae(filtered_df[road_gt],filtered_df[road_predicted])
    #print(RMSE)
    #print(MAE)
    

    plot_gt = go.Scatter(
        x = filtered_df['timestamp'],
        y = filtered_df[road_gt],
        name = 'ground_truth '
        
    )
    plot_predicted = go.Scatter(
        x = filtered_df['timestamp'],
        y = filtered_df[road_predicted],
        name = 'predicted',
        
    )
    layout = go.Layout(
        title = "prediction after 15 minutes Road {} tier-{} RMSE-{}  MAE-{}".format(road_number,road_tier_dict[int(road_number)],RMSE,MAE),
        xaxis = dict(
       
            title = 'timestamp',
            
            showgrid = True
        ),
        yaxis = dict(

            title = 'jam-length',
            showgrid = True
        )
        
        

    )
    data = [plot_gt,plot_predicted]
    fig = go.Figure(
        data = data,
        layout = layout
       
    )
    fig.update_layout(  
        template = 'plotly',
        autosize=True,
        width=1000,
        height=600,
      

        
    )
    fig.write_image('plot_pred_15_ustgcn_normal_jan_pred_after_15_day_14/{}.png'.format(title_name))
    #fig.show()
    
```

```python
for i in range(8,31,1):
    if i == 14: 
        for j in range(0,154,1):

            plot_graph(
                dataframe = ustgcn_pred_after_15,
                day = i,
                road_number = j,
                road_tier_dict =  road_tier_dict
            )

        
```
